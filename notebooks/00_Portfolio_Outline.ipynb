## Datasets

### Marathon Results (Topics 1 & 2)
- **Source:** Kaggle – 2023 Marathon Results
- **Size:** ~429,000 entries from 600+ marathon events across the United States
- **Features:** Age, Gender, Race, Finish time (seconds)
- **Usage:** Classification into performance categories (KNN) and unsupervised clustering (K-Means)
- **Reused** from Semester 1 Linear Regression project, adapted for new algorithms

### MNIST Handwritten Digits (Topic 3)
- **Source:** Built into Keras/TensorFlow (LeCun, Cortes & Burges, 1998)
- **Size:** 70,000 grayscale images (28×28 pixels)
- **Categories:** Digits 0–9
- **Usage:** Handwritten digit recognition with interactive Streamlit canvas demo

### Custom Gridworld (Topic 4)
- **Source:** Based on course materials, extended with custom features
- **Description:** 6×6 grid environment with walls/obstacles that block movement
- **Usage:** Q-Learning agent trained to navigate from start to goal while avoiding obstacles

### Fashion MNIST (Topic 5 — Planned)
- **Source:** Built into Keras/TensorFlow (Xiao, Rasul & Vollgraf, 2017)
- **Size:** 70,000 grayscale images (28×28 pixels)
- **Categories:** 10 types of clothing and accessories (T-shirt, Trouser, Pullover, Dress, Coat, Sandal, Shirt, Sneaker, Bag, Ankle boot)
- **Usage:** Product cataloguing system for e-commerce using a Convolutional Neural Network

## Planned Work

### Topic 5 — CNN (Fashion Item Classification)
**Objective:** Build a Convolutional Neural Network to classify fashion items into 10 product categories, simulating an automated product cataloguing system for e-commerce.

**Approach:**
- Load Fashion MNIST dataset from Keras
- Build a CNN architecture with convolutional layers, pooling layers, and dense layers
- Compare CNN performance against the basic ANN (Topic 3) to demonstrate why CNNs are superior for image data
- Analyse which product categories are most/least accurately classified
- Document how convolutional layers learn spatial features (edges, shapes, textures)

**Why CNN after ANN:** This demonstrates the progression from basic neural networks to deep learning. The ANN (Topic 3) treats each pixel independently, while the CNN understands spatial relationships between pixels — showing why architecture matters for image recognition.
